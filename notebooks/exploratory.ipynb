```python
# truncate all MFCC sequences to the minimum length found so that every sample has the same no. of timesteps
# Find the minimum number of timesteps across all samples
min_timesteps = min(mfcc.shape[0] for mfcc in X)
print("Minimum timesteps:", min_timesteps)

# Number of MFCC features
n_mfcc = X[0].shape[1]

# Truncate using list comprehension and stack
X_truncated = np.stack([mfcc[:min_timesteps, :] for mfcc in X])

print("Truncated X shape:", X_truncated.shape)

truncated_path = os.path.join(data_folder, "X_mfcc_truncated.npy")
np.save(truncated_path, X_truncated, allow_pickle=False)

print(f"Truncated MFCCs saved at {truncated_path}")
```

```python
import numpy as np
import os

# Load truncated MFCCs

# Correct relative path from notebooks/ folder
X = np.load("../data/X_mfcc_truncated.npy")  # shape: (3996, 1172, 20)
y = np.load("../data/y_labels.npy")          # shape: (3996,)

# Flatten each sample
X_flat = X.reshape(X.shape[0], -1)  # shape: (3996, 1172*20 = 23440)

# Normalize features
X_flat = (X_flat - X_flat.mean()) / (X_flat.std() + 1e-8)

# Save 
os.makedirs("../data", exist_ok=True)  
np.save("../data/X_flat_ready.npy", X_flat)
np.save("../data/y_labels_ready.npy", y)


print("X_flat shape:", X_flat.shape)    # (3996, 23440)
print("y shape:", y.shape)              # (3996,)
print("X_flat mean:", np.mean(X_flat))  
print("X_flat std:", np.std(X_flat))    
print("First 5 labels:",y[:5])
```

```python
import numpy as np

# LOAD EASY-TASK VAE LATENTS
Z_vae = np.load("../data/Z_vae.npy")  
print("Shape of Z_vae:", Z_vae.shape)


latent_dim = 32
num_samples = Z_vae.shape[0]

assert Z_vae.shape == (num_samples, latent_dim), \
       f"Z_vae has wrong shape. Expected (*, {latent_dim}), got {Z_vae.shape}"
print("Z_vae has correct latent dimension:", latent_dim)
```

```python
import numpy as np

# Load the one-hot genre vectors
genre_onehot = np.load("../data/genre_onehot.npy")
print("Shape of genre_onehot:", genre_onehot.shape)

# Check that each row sums to 1 (only one genre active)
row_sums = genre_onehot.sum(axis=1)
if np.all(row_sums == 1):
    print("Each row sums to 1")
else:
    print("Some rows do not sum to 1")

# Check that values are only 0 or 1
if np.all(np.logical_or(genre_onehot == 0, genre_onehot == 1)):
    print("Only 0 or 1 values in one-hot")
else:
    print("Non-binary values found")

print("First 5 one-hot vectors:\n", genre_onehot[:5])
```
